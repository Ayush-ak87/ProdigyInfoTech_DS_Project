# **Task-2**
Perform data cleaning and exploratory data analysis (EDA) on a dataset, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["Cabin"],inplace=True)
```

```python
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age,inplace=True)
df
```

```python
mode_Embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_Embarked,inplace=True)
df.info()
```

```python
df["Survived"]=df["Survived"].astype("category")
df["Pclass"]=df["Pclass"].astype("category")
df["Sex"]=df["Sex"].astype("category")
df["Age"]=df["Age"].astype("int")
df["Embarked"]=df["Embarked"].astype("category")
df.info()
```

```python
df.describe()
```

```python
sns.countplot(data=df,x="Survived")
death_percent= round(df["Survived"].value_counts(normalize=True)[0]*100)

plt.text(0, df["Survived"].value_counts()[0]+10, f"{death_percent}% died", ha="center")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

```python
#calculate passenger count by class
passenger_count= df.groupby('Pclass')['PassengerId'].count().reset_index()
passenger_count
```

```python
#plot passenger count by class
classes=['1st','2nd','3rd']
passenger_count = [216,184,491]

plt.bar(classes,passenger_count,color="blue",width=0.5)
plt.xlabel("Passenger Class")
plt.ylabel("Passenger Count")
plt.title("Passenger count by Class")
plt.show()
```

```python
#calculate and show passenger count by gender
gender_count = df['Sex'].value_counts()
plt.figure(figsize= (6,6))
plt.pie(gender_count,labels=gender_count.index,startangle=140,autopct ='% 1.1f %%')
plt.title("Distribution of Passengers by Gender")
plt.axis("equal")
plt.show()
```

```python
age_data = df['Age']
plt.figure(figsize=(8,6))
plt.hist(age_data, bins=20, edgecolor='k',color='orange')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passenger Ages')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
#Create a scatter plot
plt.figure(figsize=(17,10))
plt.scatter(df['Age'],df['Fare'],color='green',alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs.Fare')
plt.grid(True, linestyle='--',alpha=0.7)
plt.show()
```

```python
#Create a box plot
plt.figure(figsize=(7,6))
plt.boxplot([df[df['Pclass']==1]['Fare'],df[df['Pclass']== 2]['Fare'],df[df['Pclass']==3]['Fare']])
plt.title('Box Plot of Fare by Passenger Class')
plt.show()
```

```python
class_survival = df.groupby(['Pclass','Survived']).size().unstack()
print(class_survival)
#Create a stacked bar chart
class_survival.plot(kind='bar',stacked=True,color=['#ADD8E6','#000088'],alpha=1)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Status by Passenger Class')
plt.legend(['Did Not Survive','Survived'],loc='upper left')
plt.show()
```

```python
#Create a horizontal bar chart
embarkation_count= df['Embarked'].value_counts()
plt.figure(figsize=(8,6))
plt.xlabel('Count')
plt.ylabel('Embarkation Point')
embarkation_count.plot(kind='barh',color='#628ba6')
plt.title('Passenger Distribution by Embarkation Point')
plt.grid(axis='x',linestyle='--',alpha=0.7)
plt.show()
```
